```python
null_values = df.isnull().sum()
null_values
```

```python
plt.figure(figsize=(20, 10))
sns.barplot(x=null_values.index, y=null_values, color="#3894FF", edgecolor="black")
plt.title("Missing Values")
plt.xlabel("Features")
plt.ylabel("Count")
```

```python
df.drop("Cabin", axis=1, inplace=True)
```

```python
df["Age"] = df["Age"].fillna(np.mean(df["Age"]))
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df.isnull().sum()
```

```python
survived = df["Survived"]
```

```python
plt.figure(figsize=(20, 10))
sns.barplot(
    x=df["Pclass"],
    y=df["Survived"],
    color="#3894FF",
    edgecolor="white",
    err_kws={"color": "lightcoral"},
)
plt.title("Pclass vs Survived")
plt.xlabel("Pclass")
plt.ylabel("Fare")
```

```python
plt.figure(figsize=(20, 10))
sns.histplot(df, x="Age", hue="Survived", bins=10, kde=True, palette="inferno")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x=df.index, y=df["Fare"], color="#3894FF")
plt.title("Line Plot of Fare over Passenger Index")
```

```python
plt.figure(figsize=(20, 10))
sns.boxplot(data=df, x="Pclass", y="Age", color="cyan")
plt.title("Pclass vs Age")
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(
    data=df,
    x="Age",
    y="Fare",
    hue="Survived",
    edgecolor="black",
    palette=["cyan", "coral"],
)
plt.title("Age vs Fare")
```

```python
plt.figure(figsize=(20, 10))
sns.pointplot(x="Pclass", y="Survived", hue="Sex", data=df, palette=["cyan", "coral"])
plt.title("Survival Rate by Class and Gender")
plt.show()
```

```python
# Count Plot: Survival by Sex
plt.figure(figsize=(16, 9))
sns.countplot(
    x="Sex",
    hue="Survived",
    data=df,
    palette=["cyan", "coral"],
    edgecolor="black",
)
plt.title("Survival Count by Sex")
```

```python
plt.figure(figsize=(20, 10))

sns.violinplot(
    x="Survived",
    y="Fare",
    hue="Sex",
    data=df,
    split=True,
    palette=["cyan", "coral"],
    edgecolor="black",
)
plt.title("Violin Plot: Fare Distribution by Class and Gender")
plt.show()
```

```python
plt.figure(figsize=(20, 10))

sns.swarmplot(
    x="Pclass",
    y="Age",
    hue="Survived",
    data=df,
    palette=["cyan", "coral"],
    dodge=False,
    legend=False,
    size=2,
)
plt.title("Swarm Plot: Age distribution by Survival Status")
plt.show()
```

```python
plt.figure(figsize=(20, 10))

sns.kdeplot(x=df["Age"], fill=True, color="cyan")
plt.title("KDE Plot: Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.boxplot(
    data=df,
    x="Sex",
    y="Age",
    hue="Survived",
    palette=["cyan", "coral"],
)
plt.xlabel("Sex")
plt.ylabel("Age")
plt.title("Age vs Sex")
```
